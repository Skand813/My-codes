```python
import json
import wikipedia
import pandas as pd
import requests
from bs4 import BeautifulSoup
df = pd.read_excel(r'/Users/smani/Downloads/skills_taxonomy.xlsx')
skills = df["skill"].tolist()
skill_l = []
for x in skills:
    if isinstance(x,str):
        for y in x :
            if y.isalpha():
                skill_l.append(x.lower())
                break
    else:
        pass
final_skill = list(dict.fromkeys(skill_l))
final_skill = final_skill[:50]
base_url  = 'https://en.wikipedia.org/wiki/'
new_li1 = []
templ = []
for s in final_skill:
    url1 = '{}{}'.format(base_url, s)
    url2 = '{}{}{}'.format(base_url,s,"_(disambiguation)")
    resp1 = requests.get(url1)
    resp2 = requests.get(url2)
    soup1 = BeautifulSoup(resp1.text, 'lxml')
    soup2 = BeautifulSoup(resp2.text, 'lxml')
    if resp1.status_code == 200:
        for item1 in soup1.select("#firstHeading"):
            new_li1.append({"Skill":s, "URL":url1, "PageHeading":item1.text})
            print(f"Skill: " + s + " URL:"  + url1 + " Page Heading: " + item1.text )
    else :
            new_li1.append({"Skill":s, "URL":"Web site does not exist", "PageHeading":"NA"})
            print(f"Skill: " + s + " URL: " + " Web site does not exist" + " Page Heading: NA " )
    if resp2.status_code ==200:
        for item2 in soup2.select(".mw-headline"):
            templ.append(item2.getText())
            new_li1.append({"Other Uses ":templ})
        print("Other Uses :", templ)
    else :
        new_li1.append({"Other Uses ":  "NA"})
        print("Other Uses" + "NA")
            
```
