```python
import pandas as pd 
import json
import time
from PyDictionary import PyDictionary

nowtime = time.time()
f1 = open("SingleWord_" + str(nowtime) + ".json", "w+")

def google_search(i) :
    dictionary=PyDictionary()
    return dictionary.meaning(i,disable_errors=True)

df = pd.read_excel(r'/Users/smani/Downloads/Book4.xlsx')
new_list = df["CF"].tolist()
single_word = []
for items in new_list :
    spaces = items.count(" ")
    if spaces == 0 :
        single_word.append(items)
single_dict = {}
for i in single_word :
    result = google_search(i)
    single_dict["Skill"] = i
    single_dict["Google_result"]= result
    f1.write(str(single_dict)+ ",")
    
```

```python
import requests
import time
import json
import bs4
from bs4 import BeautifulSoup
import pandas as pd

nowtime = time.time()
f1 = open("Scrap_Data_total"+ str(nowtime)+".json", "w+")
test_list = [chr(x) for x in range(ord('a'), ord('z') + 1)] 

for items in test_list :
    url_n = 'https://www.consp.com/it-terminology-dictionary-'+str(items)
    #print(url_n)
    #url = 'https://www.consp.com/it-terminology-dictionary-d'
    r = requests.get(url_n,timeout=10)
    html = r.text
    soup = bs4.BeautifulSoup(html, 'lxml')
    items = soup.select(".ss-text")
    result = items[0]
    row = [i.text.replace('\n','') for i in result.select('td')]
    row_data = [col.text.strip('\n') for col in result.select('td')]
    l = row_data   
    n = 3 
    xlist = [l[i:i + n] for i in range(0, len(l), n)]  
    l1 = ["Skill(SN)","Skill(FN)","Description"]
    for i in xlist:
        final_r = dict(zip(l1,i))
        #print(final_r)
        f1.write(str(final_r) + ",") 
```
