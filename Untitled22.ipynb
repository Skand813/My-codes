```python
def num_count(lst):
    
    new_list=[]
    count = {}
    for i in lst :
        if i in count :
            count[i]+=1
            
        else :
            count[i]=1
    
    for k in range(0,101):
        #repeat = 1 
        if k in count:
            repeat = count[k]
            
            while repeat >0 :
        
                new_list.append(k)
                repeat = repeat-1
            
            
    return new_list
        
random_list= []
n = random.randint(10000,50000)
print(n)

for i in range(n):
    random_list.append(random.randint(0,101))
sortedlist = num_count(random_list) 

print(sortedlist)
```
