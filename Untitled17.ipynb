```python
import pandas as pd
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
import json
import time

df = pd.read_csv(r'/Users/smani/scrapy_projects/extra_work/driv2.csv')
links_list = df["links"].tolist()
def scraping(i) :
    resultw_dict={}
    nowtime = time.time()
    f1=open("Scrapme" + str(nowtime) + ".json", "w+")
    browser = webdriver.Firefox(executable_path = "/Users/smani/.wdm/drivers/geckodriver/macos/v0.28.0/geckodriver")
    browser.get(i)
    infothings = browser.find_elements_by_xpath('(//div[@id="ci_entry_contacts"])')
    nameschool = browser.find_element_by_xpath('(//h1[@class="wow ae-1 fromTop"])')
    resultw_dict[nameschool.text]=infothings[0].text
    f1.write(str(resultw_dict)+",")
    browser.close()
    return resultw_dict

for j in links_list[:10]:
    scraping(j)
    
```

```python
import requests
#import urllib3
import time
import json
import bs4
from bs4 import BeautifulSoup
from requests.packages.urllib3.exceptions import InsecureRequestWarning,ReadTimeoutError,HTTPError
import pandas as pd

nowtime = time.time()

pages=['list_a.html', 'list_b.html', 'list_c.html', 'list_e.html', 'list_f.html', 'list_h.html', 'list_i.html', 'list_k.html', 'list_l.html', 'list_m.html', 'list_n.html', 'list_o.html', 'list_p.html', 'list_r.html', 'list_s.html', 'list_u.html', 'list_w.html']
url = "https://www.edu.gov.mb.ca/rao/list/"
f1=open("Manitoba_New" + str(nowtime) + ".json", "w+")

for i in pages :
    url2 = "{}{}".format(url,i)
    r = requests.get(url2)
    html = r.text
    soup = bs4.BeautifulSoup(html, 'lxml')
    eleme1 = soup.select(".common")
    elements1 = eleme1[0]
    for j in elements1.select('td'):
        if j.select('p'):
            list1 =[]
            list1.append(j.text)
            continue
        if j.select('ul'):
            list2 = []
            list2.append(j.text)
            school_list=list(zip(list1,list2))
        f1.write(str(school_list)+",")
        print(school_list)
    
```
